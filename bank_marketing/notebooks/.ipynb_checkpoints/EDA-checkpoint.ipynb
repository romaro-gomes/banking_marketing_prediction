# Analyzing the distribution of numeric variables and their normalization scores using histograms and QQ plots.

```python
for x in numeric:
    sns.histplot(data=df,x=x)
    plt.title(f'Distribution of feature {x}')
    shapiro_test= shapiro(df[x])
    sm.qqplot(df[x])
    plt.title(f'Q-Q plot for {x}')
    print(f"The feature {x} has a Shapiro Test p-value of \033[1m{shapiro_test.pvalue},{'\033[1mindicating a normal distribution' if shapiro_test.pvalue > 0.05 else '\033[1mindicating a no normal distribution'}.")
    plt.show()
    print('*******************************************************************************************************************************************')
```

## Chi-Square and Cramr's Coefficient
code from [geeksforgeeks
](https://www.geeksforgeeks.org/calculate-cramer-s-coefficient-matrix-using-pandas/)

```python
def cramers_v(x, y):
    contingency_table = pd.crosstab(x, y)
    chi2_statistic, p_value, dof, expected = chi2_contingency(contingency_table)
    
    # Calculate Cramer's V
    n = contingency_table.sum().sum()
    phi2 = chi2_statistic / n
    r, k = contingency_table.shape
    phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    k_corr = k - (k - 1) * (k - 2) / (n - 1)
    r_corr = r - (r - 1) * (r - 2) / (n - 1)
    v = np.sqrt(phi2corr / min(k_corr - 1, r_corr - 1))
    
    return v
```

```python
cramers_v_matrix = pd.DataFrame(index=categories, columns=categories)

for var1 in categories:
    for var2 in categories:
        cramers_v_matrix.loc[var1, var2] = cramers_v(df[var1], df[var2])
```

```python
sns.heatmap(cramers_v_matrix.astype('float'),annot=True)
plt.title("Cramr's Coefficient per categoricals")
```

| Effect Size (ES) | Interpretation                                                                 |
|------------------|--------------------------------------------------------------------------------|
| ES  0.2         | The result is weak. Although the result is statistically significant, the fields are only weakly associated. |
| 0.2 < ES  0.6   | The result is moderate. The fields are moderately associated.                 |
| ES > 0.6         | The result is strong. The fields are strongly associated.                     |

## Mann-Whitney and Kruskal-Wallis

```python
def analyzeRelationshipNumericalAndCategorical(*,data,categorias,numericas):
    scale=len(data[categorias].unique())
    if scale ==2:
        for colname in numericas:
            list_genre=[]
            for i in range(scale):
                subset=data[(data[categorias]==data[categorias].unique()[i])][colname]
                list_genre.append(subset.tolist())
                    
            res = sp.stats.mstats.mannwhitneyu(*list_genre)
            print(f'The Mann-Whitney test evaluated whether there is a difference between the categories of {categorias} regarding {colname}. The p-value obtained was {res.pvalue:.3f}.')
            print(f'This indicates that there {"is a significant difference" if res.pvalue < 0.05 else "is no significant difference"} between the categories.')
            print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
    else:
        for colname in numericas:
            list_genre=[]
            for i in range(scale):
                subset=data[(data[categorias]==data[categorias].unique()[i])][colname]
                list_genre.append(subset.tolist())
                    
            res = sp.stats.mstats.kruskalwallis(*list_genre)
            print(f'The Kruskal-Wallis test evaluated whether there is a difference between the categories of {categorias} regarding {colname}. The p-value obtained was {res.pvalue:.3f}.')
            print(f'This indicates that there {"is a difference" if res.pvalue < 0.05 else "is no difference"} between the categories of {categorias}.')

            print('----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
        
        
```

```python
for cat in categories:
    analyzeRelationshipNumericalAndCategorical(data=df,categorias=cat,numericas=numeric)
```
